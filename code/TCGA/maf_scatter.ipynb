```python
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="ticks", font_scale=1.3)
```

```python
y = ['PTEN', 'EGFR', 'RB1', 'TP53', 'NF1',
       'PIK3R1', 'ATRX', 'PIK3CA', 'TERT', 'CIC', 'FUBP1']
```

```python
plt.rcParams['font.family'] = "Inter"
plt.figure(figsize=(20,5*len(grs)))
figcount = 1
for gr in grs:
    
    for gr_ in [0,1]:
        sub = df.loc[df.loc[:,"Neural subgroup (0=low, 1=high)"]==gr_]
        # y = sub.columns[1:].tolist()
        x = sub.index.tolist()
        plt.subplot(len(grs),2,figcount)

        for idx in tqdm(x):
            for col in y:
                color = "none"
                if sub.loc[idx, col]==gr:
                    color="black"
                plt.scatter(idx, col, color=color)

        plt.xticks([],[])
        if gr_==0:
            name="Low"
        else:
            name="High"
        
        plt.xlabel(name)
        # plt.ylabel(gr)
        if figcount%2==1:
            plt.ylabel(gr)
        figcount+=1
#plt.suptitle("Mutation variants (TCGA)")
plt.savefig("TCGA_mut.pdf")
plt.show()
```

```python
colors = sns.color_palette("colorblind", len(grs))
colormapping = dict(zip(grs, colors))
```

```python
y
```

```python
plt.rcParams['font.family'] = "Inter"
plt.figure(figsize=(24,7))

for gr_ in [0,1]:
    ax = plt.subplot(1,2,gr_+1)
    for gr in grs:
        k=0
        sub = df.loc[df.loc[:,"Neural subgroup (0=low, 1=high)"]==gr_]
        # y = sub.columns[1:].tolist()
        x = sub.index.tolist()
        for idx in tqdm(x):
            for col in y:
                color = "none"
                if sub.loc[idx, col]==gr:
                    color=colormapping[gr]
                    if k==0:
                        ax.scatter(idx, col, color=color, label=gr, marker="s", s=100)
                        k+=1
                    else:
                        ax.scatter(idx, col, color=color, label="_gr", marker="s", s=100)
                else:
                    ax.scatter(idx, col, color="none", label="_gr", marker="s", s=1)
                
        ax.set_xticks([],[])
        if gr_==0:
            name="Low"
        else:
            name="High"
        
        ax.set_xlabel(name)

        if gr_==0:
            if ax.get_legend():
                ax.get_legend().remove()
        else:
            ax.set_yticklabels([])
plt.tight_layout()
plt.legend(bbox_to_anchor=(1,1))
plt.savefig("TCGA_mut_joint.pdf", bbox_inches="tight", frameon=False)
plt.show()
```
