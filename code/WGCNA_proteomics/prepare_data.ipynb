```python
import os
import pandas as pd
```

```python
data_folder = "data"
```

```python
df = pd.read_excel(os.path.join(data_folder, "Mar23_AG_Rickleffs_GBM_Report.xlsx"), sheet_name="2", index_col=1)
df.head(10)
```

```python
rows_from = 6
cols_from = 4
colname_row = 1
```

```python
col_names = df.iloc[colname_row,cols_from:].tolist()
print(col_names[0:5])
```

```python
df_lfq = df.iloc[rows_from:, cols_from:]
df_lfq.columns = col_names
df_lfq = df_lfq.dropna()
df_lfq.head(7)
```

```python
df_lfq.index = ["batch"] + df_lfq.index[1:].tolist()
df_lfq.head(7)
```

```python
df_meta = pd.read_excel(os.path.join(data_folder, "Proteomics_Neuronal states.xlsx"), index_col=0)
```

```python
df_meta = df_meta.dropna()
```

```python
df_meta.shape
```

```python
df_lfq = df_lfq[df_meta.index.tolist()]
```

```python
print((df_lfq.columns==df_meta.index).all())
```

```python
df_meta["batch"] = df_lfq.loc["batch"].tolist()
df_lfq = df_lfq.iloc[1:,:]
```

```python
savedir = "processed"
if not os.path.exists(savedir):
    os.mkdir(savedir)
```

```python
df_meta.to_csv(os.path.join(savedir, "metadata.csv"))
```

```python
from anndata import AnnData
```

```python
df_lfq.index = [str(idx) for idx in df_lfq.index]
adata = AnnData(df_lfq.T)
adata.var_names_make_unique()
df_lfq = adata.to_df().T
```

```python
df_lfq.to_csv(os.path.join(savedir, "lfq.csv"))
```

```python
adata.obs["batch"] = df_meta.batch
```

```python
import scanpy as sc
sc.pp.scale(adata)
sc.pp.highly_variable_genes(adata)
sc.tl.pca(adata)
```
